```python
import pandas as pd
```

Loading the csv file

```python
df = pd.read_csv('/content/melb_data.csv')
df.head()
df.describe()
```

```python
df = df.dropna(axis=0)
df.describe()
```

# **Prediction Target**
Selecting the attribute which we want to predict.

```python
#selecting the target

y = df.Price
y.head()
```

# **Feature Selection**
Selecting the features which will help in predicting the target.

```python
#Selecting features

features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = df[features]
X.describe()
```

```python
X.head()
```

# **Defining a Decision Tree Model**

```python
from sklearn.tree import DecisionTreeRegressor

pricePredict_model = DecisionTreeRegressor( random_state = 1 ) #specify a no. for random state to get same results in each run
pricePredict_model.fit(X,y)  #fit the prediction target(y) and features(X) into the model
```

**Making Predictions**

```python
print("Making predictions for 1st 5 houses")
print(X.head())
print("\nThe predicted prices are : ")
print(pricePredict_model.predict(X.head()))
```

# **Mean Absolute Error MAE**



```python
from sklearn.metrics import mean_absolute_error

predicted_prices = pricePredict_model.predict(X)  #gives predicted prices of houses
mae = mean_absolute_error(y, predicted_prices) # y = actual price
print("The Mean Abosulte Error (in-sample) is : ",mae)
```

The above method is **in-sample**, which is incorrect as we are using the same data for building the model and evaluating it.

Since models' practical value come from making predictions on new data, we measure performance on data that wasn't used to build the model.

test the model's accuracy on data it hasn't seen before. This data is called validation data.

# **Splitting the data into Training and Validation data**

```python
from sklearn.model_selection import train_test_split

#splitting data into training and validation data
train_X, val_X, train_y, val_y = train_test_split(X,y,random_state=1)

#defining model
pricePredict = DecisionTreeRegressor()

#fitting model
pricePredict.fit(train_X,train_y)  #**we are using training data(part of whole data)

validation_prediction = pricePredict.predict(val_X)  #finding prediction value for val_X data
mae = mean_absolute_error(val_y,validation_prediction)
print("The Mean Absolute Error (out-sample) is : ",mae)
```

# **Underfitting and Overfitting**

**Overfitting :** capturing spurious patterns that won't recur in the future, leading to less accurate predictions.

**Underfitting :** failing to capture relevant patterns, again leading to less accurate predictions.


We are testing different values for max_leaf_nodes to see how they impact the model's accuracy, which helps in managing underfitting and overfitting. By comparing the MAE for each setting, you can select the best max_leaf_nodes value that minimizes error on the validation set without overfitting.

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    predict_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, predict_val)
    return mae

# Dictionary to store max_leaf_nodes and corresponding MAE
mae_dict = {}

for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    mae_dict[max_leaf_nodes] = my_mae
    print(f"Max leaves: {max_leaf_nodes}   \t MAE: {my_mae}")

# Find the max_leaf_nodes with minimum MAE
best_max_leaf_nodes = min(mae_dict, key=mae_dict.get)
best_mae = mae_dict[best_max_leaf_nodes]
print(f"\nBest max_leaf_nodes : {best_max_leaf_nodes} with minimum MAE : {best_mae}")
```

# **Random Forests**
The random forest uses many trees, and it makes a prediction by averaging the predictions of each component tree.

It generally has much better predictive accuracy than a single decision tree and it works well with default parameters.

We build a random forest model using "RandomForestRegressor"

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X,train_y)
prediction = forest_model.predict(val_X)

print("Mean Absolute Error : ",mean_absolute_error(val_y,prediction))
```
